```python
from openbb import obb
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

# Calculate the dynamic start_date as today minus 7 days
start_date_trailing_week = (datetime.today() - timedelta(days=7)).strftime('%Y-%m-%d')

rates = (
    obb
    .fixedincome
    .government
    .treasury_rates(
        start_date=start_date_trailing_week
    )
    .to_df()
    .dropna()
    .div(1)
    .tail(1)  # Fetch only the latest date
)
rates


# Fetch SOFR data
#sofr_data = obb.fixedincome.sofr()

# Display the data
#print(sofr_data)


#SOFR_TERM =obb.fixedincome.rate.sofr(provider='fred')
#print(SOFR_TERM)
```
